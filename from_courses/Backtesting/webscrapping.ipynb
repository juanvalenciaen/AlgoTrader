```python
data = requests.get(link3).json()
times = pd.to_datetime([e['timestamp'] for e in data])
stats = [(e['price'], e['buyLiquidations'],e['sellLiquidations'],e['openInterest'],e['openValue'],e['fundingRate'],e['fundingRateDaily'],e['indicativeFundingRate'],e['turnover'],e['turnover24h'],e['totalTurnover']) for e in data]
df = pd.DataFrame(index=times, data=stats)
df
```
